**Cauculadora co operadores**

```python
valor1 = int(input('Digite o primeiro valor: '))
valor2 = int(input('Digite o segundo valor[não pode ser zero!]: '))
valor3 = int(input('Digite o terceiro valor: '))
print(f'Soma dois valores: {valor1 + valor2}')
print(f'Soma 3 valores: {valor1 + valor2 + valor3}')
print(f'Subtração dois valores: {valor1 - valor2}')
print(f'Multiplicação dois valores: {valor1 * valor2}')
print(f'Divisão dois valores: {valor1 / valor2}')
print(f'Exponenciação dois valores: {valor1 ** valor2}')
print(f'Divisão inteira dois valores: {valor1 // valor2}')
print(f'Módulo dois valores: {valor1 % valor2}')
print(f'Média 3 valores: {(valor1 + valor2 + valor3) / 3}')
print(f'Média poderada de 5, 12, 20 e 15 compesos 1,2,3 e 4, respectivamente: {(5 * 1 + 12 * 2 + 20 * 3 + 15 * 4) / 10 }')
```

**Editando Textos**

```python
# 1. Crie uma variável chamada “frase” e atribua a ela uma string de sua escolha. Em seguida, imprima a frase na tela.
print('\n--- Tarefa 1: Frase pré-definida ---')
frase = "Esta é uma string de exemplo para a primeira tarefa."
print(frase)

# 2. Crie um código que solicite uma frase e depois imprima a frase na tela.
print('\n--- Tarefa 2: Solicitar e imprimir frase ---')
frase_digitada = input('Digite uma frase para ser impressa: ')
print(frase_digitada)

# 3. Crie um código que solicite uma frase à pessoa usuária e imprima a mesma frase digitada mas com todas as letras maiúsculas.
print('\n--- Tarefa 3: Frase em maiúsculas ---')
frase_maiuscula = input('Digite uma frase para ser impressa em maiúsculas: ')
print(frase_maiuscula.upper())

# 4. Crie um código que solicite uma frase à pessoa usuária e imprima a mesma frase digitada mas com todas as letras minúsculas.
print('\n--- Tarefa 4: Frase em minúsculas ---')
frase_minuscula = input('Digite uma frase para ser impressa em minúsculas: ')
print(frase_minuscula.lower())

# 5. Crie uma variável chamada “frase” e atribua a ela uma string de sua escolha. Em seguida, imprima a frase sem espaços em branco no início e no fim.
print('\n--- Tarefa 5: Frase pré-definida sem espaços ---')
frase_com_espacos = "   Esta é uma frase com espaços nas extremidades.   "
print(frase_com_espacos.strip())

# 6. Crie um código que solicite uma frase à pessoa usuária e imprima a mesma frase sem espaços em branco no início e no fim.
print('\n--- Tarefa 6: Solicitar e imprimir frase sem espaços ---')
frase_digitada_espacos = input('Digite uma frase com espaços (ex: "  olá mundo  "): ')
print(frase_digitada_espacos.strip())

# 7. Crie um código que solicite uma frase à pessoa usuária e imprima a mesma frase sem espaços em branco no início e no fim e em letras minúsculas.
print('\n--- Tarefa 7: Solicitar, limpar espaços e imprimir em minúsculas ---')
frase_limpa_minuscula = input('Digite uma frase (ex: "  OLÁ MUNDO  "): ')
print(frase_limpa_minuscula.strip().lower())

# 8. Crie um código que solicite uma frase à pessoa usuária e imprima a mesma frase com todas as vogais “e” trocadas pela letra “f”.
print('\n--- Tarefa 8: Trocar "e" por "f" ---')
frase_e_para_f = input('Digite uma frase para trocar as letras "e" por "f": ')
print(frase_e_para_f.replace('e', 'f').replace('E', 'F'))

# 9. Crie um código que solicite uma frase à pessoa usuária e imprima a mesma frase com todas as vogais “a” trocadas pela caractere “@”.
print('\n--- Tarefa 9: Trocar "a" por "@" ---')
frase_a_para_arroba = input('Digite uma frase para trocar as letras "a" por "@": ')
print(frase_a_para_arroba.replace('a', '@').replace('A', '@'))

# 10. Crie um código que solicite uma frase à pessoa usuária e imprima a mesma frase com todas as consoantes “s” trocadas pelo caractere “$”.
print('\n--- Tarefa 10: Trocar "s" por "$" ---')
frase_s_para_cifrao = input('Digite uma frase para trocar as letras "s" por "$": ')
print(frase_s_para_cifrao.replace('s', '$').replace('S', '$'))
```
