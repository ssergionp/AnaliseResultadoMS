```python
# Exercício 1: Maior de dois números
print("\n--- Exercício 1 ---")
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

if num1 > num2:
    print(f"O maior número é: {num1}")
elif num2 > num1:
    print(f"O maior número é: {num2}")
else:
    print("Os números são iguais.")


# Exercício 2: Crescimento/Decrescimento de produção
print("\n--- Exercício 2 ---")
percentual = float(input("Digite o percentual de crescimento de produção: "))

if percentual > 0:
    print(f"Houve um crescimento de {percentual}% na produção.")
elif percentual < 0:
    print(f"Houve um decrescimento de {abs(percentual)}% na produção.")
else:
    print("Não houve alteração no percentual de produção.")


# Exercício 3: Vogal ou Consoante
print("\n--- Exercício 3 ---")
letra = input("Digite uma letra: ").lower()
vogais = 'aeiou'

if len(letra) == 1 and letra.isalpha():
    if letra in vogais:
        print(f"A letra '{letra}' é uma vogal.")
    else:
        print(f"A letra '{letra}' é uma consoante.")
else:
    print("Por favor, digite apenas uma única letra.")


# Exercício 4: Valor mais alto e mais baixo de preços de carro
print("\n--- Exercício 4 ---")
preco1 = float(input("Digite o preço do carro no Ano 1: "))
preco2 = float(input("Digite o preço do carro no Ano 2: "))
preco3 = float(input("Digite o preço do carro no Ano 3: "))

precos = [preco1, preco2, preco3]
print(f"O valor mais alto é: R$ {max(precos):.2f}")
print(f"O valor mais baixo é: R$ {min(precos):.2f}")


# Exercício 5: Produto mais barato
print("\n--- Exercício 5 ---")
preco_prod1 = float(input("Digite o preço do Produto 1: "))
preco_prod2 = float(input("Digite o preço do Produto 2: "))
preco_prod3 = float(input("Digite o preço do Produto 3: "))

precos_produtos = [preco_prod1, preco_prod2, preco_prod3]
mais_barato = min(precos_produtos)

if mais_barato == preco_prod1:
    print(f"O Produto 1 é o mais barato (R$ {preco_prod1:.2f}).")
elif mais_barato == preco_prod2:
    print(f"O Produto 2 é o mais barato (R$ {preco_prod2:.2f}).")
else:
    print(f"O Produto 3 é o mais barato (R$ {preco_prod3:.2f}).")


# Exercício 6: Exibir 3 números em ordem decrescente
print("\n--- Exercício 6 ---")
num_a = float(input("Digite o primeiro número: "))
num_b = float(input("Digite o segundo número: "))
num_c = float(input("Digite o terceiro número: "))

numeros = [num_a, num_b, num_c]
numeros.sort(reverse=True)
print(f"Os números em ordem decrescente são: {numeros}")


# Exercício 7: Mensagem de turno de estudo
print("\n--- Exercício 7 ---")
turno = input("Em qual turno você estuda (manhã, tarde ou noite)? ").lower()

if turno == "manhã" or turno == "manha":
    print("Bom Dia!")
elif turno == "tarde":
    print("Boa Tarde!")
elif turno == "noite":
    print("Boa Noite!")
else:
    print("Valor Inválido!")


# Exercício 8: Par ou Ímpar
print("\n--- Exercício 8 ---")
numero_int = int(input("Digite um número inteiro: "))

if numero_int % 2 == 0:
    print(f"O número {numero_int} é PAR.")
else:
    print(f"O número {numero_int} é ÍMPAR.")


# Exercício 9: Inteiro ou Decimal
print("\n--- Exercício 9 ---")
numero_check = float(input("Digite um número: "))

if numero_check == int(numero_check):
    print(f"O número {numero_check} é INTEIRO.")
else:
    print(f"O número {numero_check} é DECIMAL.")


# Exercício 10 (Projeto): Operação e propriedades do número
print("\n--- Exercício 10 (Projeto) ---")
num_p1 = float(input("Digite o primeiro número: "))
num_p2 = float(input("Digite o segundo número: "))
operacao = input("Qual operação deseja realizar (+, -, *, /, //, %)? ")

resultado = 0
if operacao == '+':
    resultado = num_p1 + num_p2
elif operacao == '-':
    resultado = num_p1 - num_p2
elif operacao == '*':
    resultado = num_p1 * num_p2
elif operacao == '/':
    if num_p2 != 0:
        resultado = num_p1 / num_p2
    else:
        print("Erro: Divisão por zero!")
        resultado = None # Indica erro
elif operacao == '//':
    if num_p2 != 0:
        resultado = num_p1 // num_p2
    else:
        print("Erro: Divisão inteira por zero!")
        resultado = None
elif operacao == '%':
    if num_p2 != 0:
        resultado = num_p1 % num_p2
    else:
        print("Erro: Módulo por zero!")
        resultado = None
else:
    print("Operação inválida.")
    resultado = None

if resultado is not None:
    print(f"O resultado da operação é: {resultado}")

    # Propriedades do resultado
    if resultado == int(resultado):
        print("O resultado é INTEIRO.")
        if int(resultado) % 2 == 0:
            print("O resultado é PAR.")
        else:
            print("O resultado é ÍMPAR.")
    else:
        print("O resultado é DECIMAL.")

    if resultado > 0:
        print("O resultado é POSITIVO.")
    elif resultado < 0:
        print("O resultado é NEGATIVO.")
    else:
        print("O resultado é ZERO.")


# Exercício 11 (Projeto): Tipos de Triângulo
print("\n--- Exercício 11 (Projeto) ---")
lado1 = float(input("Digite o comprimento do primeiro lado: "))
lado2 = float(input("Digite o comprimento do segundo lado: "))
lado3 = float(input("Digite o comprimento do terceiro lado: "))

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("Os valores podem formar um triângulo.")
    if lado1 == lado2 == lado3:
        print("É um triângulo EQUILÁTERO.")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("É um triângulo ISÓSCELES.")
    else:
        print("É um triângulo ESCALENO.")
else:
    print("Os valores NÃO podem formar um triângulo.")


# Exercício 12 (Projeto): Cálculo de combustível com desconto
print("\n--- Exercício 12 (Projeto) ---")
litros = float(input("Digite a quantidade de litros vendidos: "))
tipo_combustivel = input("Digite o tipo de combustível (E para Etanol, D para Diesel): ").upper()

preco_litro = 0
desconto_percentual = 0

if tipo_combustivel == 'E':
    preco_litro = 1.70
    if litros <= 15:
        desconto_percentual = 0.02 # 2%
    else:
        desconto_percentual = 0.04 # 4%
elif tipo_combustivel == 'D':
    preco_litro = 2.00
    if litros <= 15:
        desconto_percentual = 0.03 # 3%
    else:
        desconto_percentual = 0.05 # 5%
else:
    print("Tipo de combustível inválido.")
    preco_litro = 0 # Para evitar cálculo com valor incorreto

if preco_litro > 0:
    valor_bruto = preco_litro * litros
    valor_desconto = valor_bruto * desconto_percentual
    valor_a_pagar = valor_bruto - valor_desconto
    print(f"O valor a ser pago pelo cliente é: R$ {valor_a_pagar:.2f}")


# Exercício 13 (Projeto): Análise de vendas anuais
print("\n--- Exercício 13 (Projeto) ---")
vendas_2022 = float(input("Digite a quantidade de vendas em 2022: "))
vendas_2023 = float(input("Digite a quantidade de vendas em 2023: "))

if vendas_2022 == 0:
    if vendas_2023 > 0:
        print("Houve vendas em 2023, mas não em 2022. Não é possível calcular variação percentual.")
    else:
        print("Não houve vendas em 2022 nem em 2023.")
else:
    variacao_percentual = ((vendas_2023 - vendas_2022) / vendas_2022) * 100
    print(f"Variação percentual de vendas: {variacao_percentual:.2f}%")

    if variacao_percentual > 20:
        print("Sugestão: Bonificação para o time de vendas.")
    elif 2 <= variacao_percentual <= 20:
        print("Sugestão: Pequena bonificação para time de vendas.")
    elif -10 < variacao_percentual < 2:
        print("Sugestão: Planejamento de políticas de incentivo às vendas.")
    else: # variacao_percentual <= -10
        print("Sugestão: Corte de gastos.")
```
